```python
import json
```

```python
def load_data(filename):
    with open(filename, "r") as f:
        data =  json.load(f)
        
    return data
```

```python
data = load_data("store_data.json")
```

```python
print(data)
print(type(data))
```

# clean and structured the data

```python
def clean_data(data):
    text_to_number = {"one": 1, "two": 2, "three" : 3, "four": 4, "five": 5}
    cleaned_data = []
    unique_user = set()
    print(unique_user)
    
    for user in data:
        # clean rating
        raw_rating = str(user["rating"]).strip().lower()

        if raw_rating in text_to_number:
            user["rating"] = text_to_number[raw_rating]

        # handle missing age
        raw_age = user.get("age")
        if raw_age is None:
            user["age"] = None

        # deduplication
        if(user["name"].strip() in unique_user):
            continue;

        unique_user.add(user["name"])
        cleaned_data.append(user)

    return cleaned_data;
    
```

```python
clean_data(data)
```

```python
def get_insights(data):
    #avg rating
    total_rating = 0
    for user in data:
        rating = user.get("rating")

        if rating is None:
            continue;
            
        else:
            total_rating += float(rating)

    print(f"avg rating : {total_rating}")
    
```

```python
get_insights(data)
```
